```python
# part 1
with open("input.txt") as file:
    lines = file.readlines()

paperList = [[]]

for line in lines:
    for char in line:
        if char == "\n":
            paperList.append([])
        else:
            paperList[-1].append(char)

totalRolls = 0

for row in range(len(paperList)):
    for col in range(len(paperList[row])):
        if paperList[row][col] != "@":
            continue
        surroundings=0
        # not left bounded, so check everything to the left
        if col != 0:
            if paperList[row][col-1] == "@":
                surroundings += 1
            if row !=0:
                if paperList[row-1][col-1] == "@":
                    surroundings += 1
            if row+1 < len(paperList) and col < len(paperList[row+1]):
                if paperList[row+1][col-1] == "@":
                    surroundings += 1
        # not right bounded, so check everything to the right
        if col != len(paperList[row]) - 1:
            if paperList[row][col+1] == "@":
                surroundings += 1
            if row !=0:
                if paperList[row-1][col+1] == "@":
                    surroundings += 1
            if row+1 < len(paperList) and col < len(paperList[row+1]):
                if paperList[row+1][col+1] == "@":
                    surroundings += 1
        # not top or bottom bounded, so check up and down
        if row != 0:
            if paperList[row-1][col] == "@":
                surroundings += 1
        if row+1 < len(paperList) and col < len(paperList[row+1]):
            if paperList[row+1][col] == "@":
                surroundings += 1
        if surroundings < 4:
            print(f"row: {row}, col: {col}, surroundings: {surroundings}")
            totalRolls += 1

print(totalRolls)
```

```python
# part 2
with open("input.txt") as file:
    lines = file.readlines()

paperList = [[]]

for line in lines:
    for char in line:
        if char == "\n":
            paperList.append([])
        else:
            paperList[-1].append(char)

totalRolls = 0


def checkList(totalRolls):
    numCellsUpdated = 0
    for row in range(len(paperList)):
        for col in range(len(paperList[row])):
            if paperList[row][col] != "@":
                continue
            surroundings=0
            # not left bounded, so check everything to the left
            if col != 0:
                if paperList[row][col-1] == "@":
                    surroundings += 1
                if row !=0:
                    if paperList[row-1][col-1] == "@":
                        surroundings += 1
                if row+1 < len(paperList) and col < len(paperList[row+1]):
                    if paperList[row+1][col-1] == "@":
                        surroundings += 1
            # not right bounded, so check everything to the right
            if col != len(paperList[row]) - 1:
                if paperList[row][col+1] == "@":
                    surroundings += 1
                if row !=0:
                    if paperList[row-1][col+1] == "@":
                        surroundings += 1
                if row+1 < len(paperList) and col < len(paperList[row+1]):
                    if paperList[row+1][col+1] == "@":
                        surroundings += 1
            # not top or bottom bounded, so check up and down
            if row != 0:
                if paperList[row-1][col] == "@":
                    surroundings += 1
            if row+1 < len(paperList) and col < len(paperList[row+1]):
                if paperList[row+1][col] == "@":
                    surroundings += 1
            if surroundings < 4:
                print(f"row: {row}, col: {col}, surroundings: {surroundings}")
                totalRolls += 1
                paperList[row][col]="."
    return totalRolls

while True:
    rolls = checkList(totalRolls)
    if rolls == totalRolls:
        break
    totalRolls = rolls

print(totalRolls)
```
