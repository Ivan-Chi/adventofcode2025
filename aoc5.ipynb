```python
#Read data
with open('aoc5input.txt') as f:
    inputStr=f.read()
    
#Process Data
grid=inputStr.split('\n')
i=grid.index('')
rangeList,ids=grid[:i], grid[i+1:]

# Sort ranges
sortedRangeList = [range.split('-') for range in rangeList]
sortedRangeList.sort(key=lambda x: int(x[0]))

mergedRangeList=[[]]

# Build merged range list
i=0
for i in range(len(sortedRangeList)):
    curr_start=int(sortedRangeList[i][0])
    curr_end=int(sortedRangeList[i][1])

    if i==0:
        mergedRangeList[0]=[curr_start,curr_end]
        continue

    last_end=mergedRangeList[-1][1]
    
    # No overlap case
    if curr_start > last_end:
        mergedRangeList.append([curr_start,curr_end])
        continue

    # Overlap case
    if curr_end > last_end:
        mergedRangeList[-1][1]=curr_end
    
numFresh=0
# ID search
for id in ids:
    if not id:
        continue
    id_int=int(id)
    for r in mergedRangeList:
        if r[0]<=id_int<=r[1]:
            numFresh+=1
            break

print(numFresh)

        


    
```
